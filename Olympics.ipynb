```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
athlete = pd.read_excel("olympics_dataset/Athletes.xlsx")
coach = pd.read_excel("olympics_dataset/Coaches.xlsx")
gender = pd.read_excel("olympics_dataset/entriesgender.xlsx")
medals = pd.read_excel("olympics_dataset/medals.xlsx")
teams = pd.read_excel("olympics_dataset/Teams.xlsx")
```

# Medal Distribution


## Top 10 countries with highest number of medals (total)

```python
top_10 = medals.sort_values('Rank by Total').head(10)
plt.bar(top_10['Team/NOC'], top_10['Total'], width=0.8, color='Pink')
plt.title('TOP 10 COUNTRIES WITH HIGHEST TOTAL MEDALS', fontsize=20)
plt.xlabel('Countries')
plt.xticks(rotation=90)
plt.ylabel('No. of medals')
plt.show()
```

## Top 10 countries with highest number of gold

```python
top_10_gold = medals.sort_values('Gold', ascending=False).head(10)
plt.bar(top_10_gold['Team/NOC'], top_10_gold['Gold'], width=0.8, color='Gold')
plt.title('TOP 10 COUNTRIES WITH HIGHEST GOLD MEDAL', fontsize=20)
plt.xlabel('Countries')
plt.xticks(rotation=90)
plt.ylabel('No. of golds')
plt.show()
```

## Top 10 countries with highest number of silver

```python
top_10_silv = medals.sort_values('Silver', ascending=False).head(10)
plt.bar(top_10_silv['Team/NOC'], top_10_silv['Silver'], width=0.8, color='Silver')
plt.title('TOP 10 COUNTRIES WITH HIGHEST SILVER MEDAL', fontsize=20)
plt.xlabel('Countries')
plt.xticks(rotation=90)
plt.ylabel('No. of Silvers')
```

## Top 10 countries with highest number of bronze

```python
top_10_bron = medals.sort_values('Bronze', ascending=False).head(10)
plt.bar(top_10_bron['Team/NOC'], top_10_bron['Bronze'], width=0.8, color='Brown')
plt.title('TOP 10 COUNTRIES WITH HIGHEST BRONZE MEDAL', fontsize=20)
plt.xlabel('Countries')
plt.xticks(rotation=90)
plt.ylabel('No. of Bronze')
plt.show()
```

## Overall medal distribution

```python
fig = plt.figure()
fig.set_facecolor('White')
plt.pie(medals['Total'], labels=medals['Team/NOC'], radius=5, rotatelabels=90)
plt.show()
```

# Athlete Analysis

## Countries with highest number of athletes

```python
num_ath = athlete.groupby('NOC').size()
num_ath=num_ath.to_frame('Count') 
num_ath=num_ath.sort_values('Count', ascending=False).head(15)
plt.bar(num_ath.index.values, num_ath['Count'])
plt.title('COUNTRIES WITH THE HIGHEST NUMBER OF ATHLETES', fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Athlete Count')
plt.xticks(rotation=90)
plt.show()
```

## Overall athlete distribultion

```python
fig = plt.figure()
fig.set_facecolor('White')
tot_ath = athlete.groupby('NOC').size()
tot_ath=tot_ath.to_frame('Count') 
tot_ath=tot_ath.sort_values('Count', ascending=False)
plt.pie(tot_ath['Count'], labels=tot_ath.index.values, radius=5, rotatelabels=90)
plt.show()
```

# Gender Analysis

## Female to male distribution

```python
fem = sum(gender['Female'])
male = sum(gender['Male'])
series = [fem, male]
names=['Female', 'Male']
fig=plt.figure()
fig.set_facecolor('White')
plt.pie(series, labels=names, wedgeprops={'edgecolor':'black', 'linewidth': '2'}, autopct='%1.1f%%')
plt.show()
```

## Discipline with the highest number of females

```python
fem_top_event = gender.sort_values('Female', ascending=False).head(10)
plt.bar(fem_top_event['Discipline'], fem_top_event['Female'], width=0.8, color='Red')
plt.title('TOP 10 DISCIPLINES WITH THE HIGHEST FEMALE COUNT', fontsize=20)
plt.xlabel('Countries')
plt.xticks(rotation=90)
plt.ylabel('Female count')
plt.show()
```

## Discipline with the highest number of Men

```python
men_top_event = gender.sort_values('Male', ascending=False).head(10)
plt.bar(men_top_event['Discipline'], men_top_event['Male'], width=0.8, color='Cyan')
plt.title('TOP 10 DISCIPLINES WITH THE HIGHEST MALE COUNT', fontsize=20)
plt.xlabel('Countries')
plt.xticks(rotation=90)
plt.ylabel('Male count')
plt.show()
```

# Coach Analysis

## Countries with the highest number of coaches

```python
num_coa = coach.groupby('NOC').size()
num_coa=num_coa.to_frame('Count') 
num_coa=num_coa.sort_values('Count', ascending=False).head(15)
plt.bar(num_coa.index.values, num_coa['Count'], color='Orange')
plt.title('COUNTRIES WITH THE HIGHEST NUMBER OF COACHES', fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Coach Count')
plt.xticks(rotation=90)
plt.show()
```

## Countries with the lowest number of coaches

```python
low_coa = coach.groupby('NOC').size()
low_coa=low_coa.to_frame('Count') 
low_coa=low_coa.sort_values('Count').head(15)
plt.bar(low_coa.index.values, low_coa['Count'], color='Black')
plt.title('COUNTRIES WITH THE LOWEST NUMBER OF COACHES', fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Coach Count')
plt.xticks(rotation=90)
plt.show()
```

# Number of Teams (Female+Male)

```python
num_team = teams.groupby('Name').size()
num_team=num_team.to_frame('Count') 
num_team=num_team.sort_values('Count', ascending=False).head(15)
plt.bar(num_team.index.values, num_team['Count'], color='Purple')
plt.title('COUNTRIES WITH THE HIGHEST NUMBER OF TEAMS', fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Team Count')
plt.xticks(rotation=90)
plt.show()
```
